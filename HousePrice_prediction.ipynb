```python
#To slove the following mess with the data we can use the log transformation technique to make the data normally distributed 
# 1.crime_rate,dist_1,dist_2,dist3,dist4 is left skewed
# 2.age column data is right skewed 
df['crime_rate'] = np.log(df['crime_rate']+1) 
df['dist1'] = np.log(df['dist1'] + 1) 
df['dist2'] = np.log(df['dist2'] + 1)  
df['dist3'] = np.log(df['dist3'] + 1) 
df['dist4'] = np.log(df['dist4'] + 1)  
df['age'] = np.log(df['age'])
```

```python
#To slove the outliers problem in the dataset using the IQR method 
# 3.resid_area has some outliers 4.air_qual also contain some outlier
Q1_1 = df['resid_area'].quantile(0.25)
Q3_1 = df['resid_area'].quantile(0.75)
IQR = Q3_1 - Q1_1

lower_bound_1 = Q1_1 - 1.5 * IQR
upper_bound_1 = Q3_1 + 1.5 * IQR

df = df[(df['resid_area'] >= lower_bound_1) & (df['resid_area'] <= upper_bound_1)] 
```

```python
Q1 = df['air_qual'].quantile(0.25)
Q3 = df['air_qual'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['air_qual'] >= lower_bound) & (df['air_qual'] <= upper_bound)]
```

```python
#Importing the pipeline to start the building the model 
from imblearn.pipeline import Pipeline 
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OneHotEncoder
```

```python
#Creating the pipeline for the categorical column  
steps_1 = [('Imputer',SimpleImputer(fill_value='missing',strategy='constant')),('encoding',OneHotEncoder())]
categorical_preproccesor = Pipeline(steps_1)
```

```python
categorical_preproccesor
```

```python
#Creating the pipeline for the numerical columns 
from sklearn.preprocessing import StandardScaler
steps_2 = [('imputrt_median',SimpleImputer(missing_values=np.nan,strategy='mean')),('Scalar',StandardScaler())] 
numerical_preprocessor = Pipeline(steps_2)
```

```python
numerical_preprocessor 
```

```python
df.dtypes
```

```python
#Importing the column trnsformer to combine the numerical and categorical preprocessors 
from sklearn.compose import ColumnTransformer 
categorical_columns = ['airport','waterbody','bus_ter'] 
numerical_columns = ['crime_rate','resid_area','air_qual','room_num','age','dist1','dist2','dist3','dist4','teachers','poor_prop','n_hos_beds','n_hot_rooms','rainfall','parks']
```

```python
preprocessor = ColumnTransformer(transformers=[('categorical',categorical_preproccesor,categorical_columns) , ('numerical',numerical_preprocessor,numerical_columns)])
```

```python
preprocessor
```

```python
#Splitting the features and the target variable 
X = df.drop('price', axis=1)
y = df['price'] 
```

```python
#Spliting the data into train and test 
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state = 42)
```

```python
#Importing the model for the testing 
from sklearn.linear_model import LinearRegression
```

```python
#Creating a pipeline with preprocessingand the model 
steps = [('preprocessing',preprocessor),('Regression',LinearRegression())]
model_1 = Pipeline(steps)
```
