```python
#Importing the pipeline to create the pipeline to fix the nulls in the following columns 
from imblearn.pipeline import Pipeline 
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import StandardScaler
```

```python
#Pipeline to fix the null values 
steps_1 = [('Imputer',SimpleImputer(strategy='median')),('Scalar',StandardScaler())]
numerical_nulls_preprocessor = Pipeline(steps_1) 
```

```python
numerical_nulls_preprocessor
```

```python
#Pipeline to fix the nulls in the categorical columns 
from sklearn.preprocessing import OneHotEncoder
steps_2 = [('Imputer_constant',SimpleImputer(fill_value='Unknown',strategy='constant')),('Encoder',OneHotEncoder())] 
categorical_nulls_preprocessor = Pipeline(steps_2)
```

```python
categorical_nulls_preprocessor
```

```python
df_train.dtypes
```

```python
#Importing the column transformer to combine these two pipelines
from sklearn.compose import ColumnTransformer
```

```python
numerical_columns = ['Age','Annual_Income','Monthly_Inhand_Salary','Num_Bank_Accounts','Num_Credit_Card','Interest_Rate','Num_of_Loan','Delay_from_due_date','Num_of_Delayed_Payment','Changed_Credit_Limit','Num_Credit_Inquiries','Outstanding_Debt','Credit_Utilization_Ratio','Credit_History_Age','Total_EMI_per_month','Amount_invested_monthly','Monthly_Balance']
```

```python
categorical_columns = ['Occupation','Type_of_Loan','Credit_Mix','Payment_of_Min_Amount','Payment_Behaviour']
```

```python
preprocessor = ColumnTransformer(transformers=[('numerical',numerical_nulls_preprocessor,numerical_columns) , ('categorical',categorical_nulls_preprocessor,categorical_columns)]) 
```

```python
preprocessor
```

```python
#Splitting the data 
X = df_train.iloc[:,:-1] 
y= df_train.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
#Creating a model Pipeline 
from sklearn.linear_model import LogisticRegression 
from imblearn.over_sampling import SMOTE
```

```python
steps_6 = [('preprocessing',preprocessor),('smote',SMOTE()),('Classifer',LogisticRegression())] 
model_1 = Pipeline(steps_6)
```

```python
model_1.fit(X_train,y_train)
```

```python
from sklearn.model_selection import RandomizedSearchCV

grid = {
    'Classifer__C': [0.01, 0.1, 1, 10],
    'Classifer__penalty': ['l1','l2'],
    'Classifer__solver': ['liblinear','saga'],
    'Classifer__max_iter': [100, 200]
}


rs = RandomizedSearchCV(
    estimator=model_1,
    param_distributions=grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    n_iter=12,
    random_state=42,
    verbose=2
)

rs.fit(X_train, y_train)

print(rs.best_params_)
print(rs.best_score_)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import make_scorer, f1_score, roc_auc_score

grid = {
    'Classifer__C': [0.05, 0.1, 0.2, 0.5],
    'Classifer__max_iter': [150, 200, 300]
}
scoring = {
    'f1': make_scorer(f1_score, average='macro'),
    'roc_auc': make_scorer(roc_auc_score, needs_proba=True, multi_class='ovr', average='macro')
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

gs = GridSearchCV(
    estimator=model_1,
    param_grid=grid,
    scoring=scoring,
    refit='f1',
    cv=cv,
    n_jobs=-1,
    verbose=2
)

gs.fit(X_train, y_train)

print("Best Params:", gs.best_params_)
print("Best F1 (macro):", gs.best_score_)
print("Best ROC-AUC (macro):", gs.cv_results_['mean_test_roc_auc'][gs.best_index_])
```
